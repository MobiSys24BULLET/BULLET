```python
import pandas as pd
import glob
import numpy as np
```

# Calculate TP/FP for all samples

```python
# Set a threshold to determine if the Proactive BSR value is considered a positive or negative estimation
threshold = 0

# Find all bullet_trig_evalN.csv files
files = []
for i in range(2):
    filename = 'bullet_trig_eval{}.csv'.format(i)
    if glob.glob(filename):
        files.append(filename)

results = []


# Process each file
for file in files:
    df = pd.read_csv(file, delimiter='\t') 
    df = df[df['Time'] > 0.4]

    # Calculate TP, FP, TN, and FN
    TP = ((df["True Buffer"] > 0) & (df["Total BSR"] > threshold)).sum()
    FP = ((df["True Buffer"] == 0) & (df["Total BSR"] > threshold)).sum()
    TN = ((df["True Buffer"] == 0) & (df["Total BSR"] <= threshold)).sum()
    FN = ((df["True Buffer"] > 0) & (df["Total BSR"] <= threshold)).sum()

    # Calculate true positive rate (TPR) and false positive rate (FPR)
    TPR = TP / (TP + FN)
    FPR = FP / (FP + TN)

    results.append({
        'file': file,
        'True Positive Rate': TPR,
        'False Positive Rate': FPR
    })

# Convert the results to a DataFrame and display
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# Initialize the error counter and row counter
error_sum = 0
row_count = 0
error_list = []
import matplotlib.pyplot as plt

for file in files:
    df = pd.read_csv(file, delimiter='\t')
    df = df[df['Time'] > 0.4 + 0.05*i] # scheduling slot
    df = df[df['Time']*1000%2.5 == 0.5] # scheduling slot
    
    #df = df[df['BSR'] == 0]
    
    mode_counts = df["Mode"].value_counts()
    num_zeros = mode_counts.get(0, 0)
    num_ones = mode_counts.get(1, 0)
    mode_ratio = num_ones / (num_zeros + num_ones + 1)

    print(f"File: {file}")
    print(f"Number of 1's in 'Mode' column: {num_ones}")
    print(f"Number of 0's in 'Mode' column: {num_zeros}")
    print("Ratio of 1 to total (1 and 0) in 'Mode' column:", mode_ratio)

    # Calculate the absolute error between True Buffer and Proactive BSR
    diff = df["True Arrival"] - df["Estimated_Arrival"] 
    diff = df["True Buffer"] - df["Total BSR"] 
    abs_error = np.abs(diff)
    
    plt.plot(df['Time'], diff)
    plt.show()
    
    # Append the absolute error to the error list
    error_list.extend(abs_error)

    # Update the error counter and row counter
    error_sum += abs_error.sum()
    row_count += len(df)
    print('Error: ', abs_error.sum()/len(df))

# Calculate the average absolute error
average_abs_error = error_sum / row_count

# Calculate the standard deviation of the error
std_error = np.std(error_list)

print("Average Absolute Error:", average_abs_error, np.percentile(error_list, 12.5),np.percentile(error_list, 87.5))
print("Standard Deviation of Absolute Error:", std_error)
```

```python
import csv
import numpy as np

# Open the CSV file and read the data into a list
with open('bullet_trig_eval3.csv', 'r') as file:
    reader = csv.reader(file, delimiter='\t')
    data = [row for row in reader]

# Convert the data to a NumPy array and extract the 'Arrival' column
data = np.array(data[1:], dtype=np.float64)
arrival = data[:, 3][10000:22000]

# Compute the FFT of the 'Arrival' column
fft_result = np.fft.fft(arrival)

# Compute the magnitude of the FFT
fft_magnitude = np.abs(fft_result)
fft_magnitude[0] = 0

# Compute the sample rate and length of the FFT result
sample_rate = 2000  # Replace with the actual sample rate of the data
fft_length = len(fft_result)

# Find the maximum value in the magnitude array
max_magnitude = np.max(fft_magnitude)

# Compute the corresponding frequency
frequency = np.where(fft_magnitude == max_magnitude)[0][0] / fft_length * sample_rate

# Print the result
print(f"Maximum frequency value: {frequency}")
plt.plot(fft_magnitude)
plt.xlim([0,100])
```
