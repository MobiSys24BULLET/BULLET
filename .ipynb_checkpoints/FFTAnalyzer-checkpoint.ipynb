```python
import os
import numpy as np
import matplotlib.pyplot as plt

for i in range(0, 11):
    file = f'bullet_trig_fft{i}.csv'

    if not os.path.exists(file):
        print(f"File {file} not found.")
        continue

    # Initialize empty lists for time and max_values
    time_list = []
    max_values_list = []
    top_5_values_list = []

    # Read the file line by line
    with open(file, 'r') as f:
        for line in f:
            # Split the elements by tab
            elements = line.strip().split('\t')

            # Extract time (first element) and convert it to float
            time = float(elements[0])

            # Convert the remaining elements to float
            float_elements = list(map(float, elements[1:]))

            # Find the index of the maximum value (excluding time)
            max_value_index = float_elements.index(max(float_elements))
            
            # Find the indices of the top-5 values (excluding time)
            top_5_indices = 1000/(np.argsort(float_elements)[-5:] + 1)

            
            # Append the time and max_value to the lists
            time_list.append(time)
            max_values_list.append(max_value_index)
            top_5_values_list.append(top_5_indices)

    # Plot the graph
    plt.figure()
    plt.scatter(time_list, max_values_list, color='blue', label='Max values')

    # Plot top-5 values with different colored dots
    colors = ['red', 'green', 'orange', 'purple', 'brown']
    for t, top_5 in zip(time_list, top_5_values_list):
        for i, value in enumerate(top_5):
            plt.scatter(t, value, color=colors[i], label=f'Top {i+1}' if t == time_list[0] else "")

    plt.xlabel('Time')
    plt.ylabel('Values')
    plt.ylim([0,100])
    plt.title(f'Time vs Max and Top-5 values for {file}')
    plt.legend(['1', '2', '3', '4', '5'])
    #plt.savefig(f'plot_{file[:-4]}.png')  # Save the figure as an image
    #plt.close()  # Close the figure to free memory
```
