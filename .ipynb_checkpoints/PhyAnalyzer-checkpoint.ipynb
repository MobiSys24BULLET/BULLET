```python
import pandas as pd
import matplotlib.pyplot as plt

# read the csv file
df = pd.read_csv('RxPacketTrace.txt', delimiter='\t')


# Group the data by rnti
grouped = df.groupby('rnti')

# Plot the graph for each rnti grouped file
for rnti, group in grouped:
    # Create a new figure
    fig, ax1 = plt.subplots()

    # Plot mcs on the left y-axis
    ax1.plot(group["Time"], group["mcs"], color="blue")
    ax1.set_xlabel("Time")
    ax1.set_ylabel("MCS", color="blue")
    ax1.tick_params(axis="y", labelcolor="blue")

    # Create a second y-axis on the right
    ax2 = ax1.twinx()

    # Plot SINR(dB) on the right y-axis
    ax2.plot(group["Time"], group["SINR(dB)"], color="red")
    ax2.set_ylabel("SINR(dB)", color="red")
    ax2.tick_params(axis="y", labelcolor="red")

    # Add a title and legend
    ax1.set_title(f"RNTI {rnti}")
    ax1.legend(["MCS"], loc="upper left")
    ax2.legend(["SINR(dB)"], loc="upper right")
    
    print('MCS: ', np.mean(group["mcs"], np.std(group["mcs"], 'SINR (dB): ', np.mean(group["SINR (dB)"], np.std(group["SINR (dB)"])

    # Display the plot
    plt.show()
```
